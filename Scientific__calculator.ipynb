```python
import math
import numpy as np
```

```python
import numpy as np
def add(a, b):
    return a + b

def subtract(a , b):
    return a - b

def multiply(a , b):
    return a * b

def divide(a , b):
    return a / b

def exp(x):
    result = math.sqrt(x)
    return result 

def sin(x):
    result = math.sin(x)
    return result

def cos(x):
    result = math.cos(x)
    return result 

def tan(x):
    result = math.tan(x)
    return result

def arcsin(x):
    result = np.asin(x)
    return result

def acrcos(x):
    result = np.acos(x)
    return result

def arctan(x):
    result = math.atan(x)
    return result

def radians(x):
    result = math.radians(x)
    return result 

def degree(x):
    result = math.degree(x)
    return result

def exponent(x):
    result = math.exp(x)
    return result

def naturallog(x):
    result = math.log(x)
    return result 

def base10log(x):
    result = math.log10(x)
    return result 

def power(x):
    result = math.pow(x)
    return result

def mean(x):
    result = math.mean(x)
    return result

def median(x):
    result = math.median(x)
    return result

def mode(x):
    result = math.mode(x)
    return result




print("""
Choose a number for the following operations
1 - Addition(x,y)
2 - Subtraction(x,y)
3 - Multiplication(x,y)
4 - Divide(x,y)
5 - Square(x)
6 - Sin(x)
7 - Cos(x)
8 - Tan(x)
9 - Arcsin(x)
10 - Arccos(x)
11 - Arctan(x)
12 - Radians(x)
13 - Degree(x)
14 - Exponent(x)
15 - Naturallog(x)
16 - Base10log(x)
17 - Power(x)
18 - Mean(x)
19 - Median(x)
20 - Mode(x)""")


op = int(input('What operation do you want to perform?'))
while op < 23:
    if op == 1:
        print("Enter the parameters")
        x1 = int(input("enter x"))
        y1 = int(input("enter y"))
        res1 = add(x1,y1)
        print("Addition= ", res1)

    elif op == 2:
        x2 = int(input("enter x"))
        y2 = int(input("enter y"))
        res2 = subtract(x2,y2)
        print("Subtraction= ", res2)

    elif op == 3:
        x3 = int(input("enter x"))
        y3 = int(input("enter y"))
        res3 = multiply(x3,y3)
        print("Multiplication= ", res3)

    elif op == 4:
        x4 = int(input("enter x"))
        y4 = int(input("enter y"))
        res4 = divide(x4,y4)
        print("Division= ", res4)
    elif op == 5:
        x5 = int(input("enter x"))
        res5 = exp(x5)
        print("Square= ", res5)

    elif op == 6:
        x6 = int(input("enter x"))
        res6 = sin(x6)
        print("sin(x)= ", res6 )

    elif op == 7:
        x7 = int(input("enter x"))
        res7 = cos(x7)
        print("cos(x)= ", res7)

    elif op == 8:
        x8 = int(input("enter x"))
        res8 = tan(x8)
        print("tan(x)= ", res8)

    elif op == 9:
        x9 = int(input("enter x"))
        res9 = np.arcsin(x9)
        print("arcsin(x)= ", res9)

    elif op == 10:
        x10 = int(input("enter x"))
        res10 = np.arccos(x10)
        print("arccos(x)= ", res10)

    elif op == 11:
        x11 = int(input("enter x"))
        res11 = arctan(x11)
        print("arctan(x)= ", res11)

    elif op == 12:
        x12 = int(input("enter x"))
        res12 = radians(x12)
        print("radian(x)= ", res12)

    elif op == 13:
        x13 = int(input("enter angle in degrees"))
        res13 =math.radians(x13)
        print("degree(x)= ", res13)


    elif op == 14:
        x14 = int(input("enter x"))
        from math import exp
        res14 = exp(x14)
        print("exp(x)= ", res14)

    elif op == 15:
        x15 = int(input("enter x"))
        res15 = naturallog(x15)
        print("naturallog(x)= ", res15)

    elif op == 16:
        x16 = int(input("enter x"))
        res16 = base10log(x16)
        print("base10log(x)= ", res16)

    elif op == 17:
        x17_a = int(input("enter base"))
        x17_b = int(input("enter power"))
        res17 = x17_a ** x17_b
        print("power(x)= ", res17)

    elif op == 18:
        temp = []
        # x18 = 
        # res18 = mean(*x18)
        n = int(input("How many numbers you want to use to calculate mean?"))
        for i in range(n):
            temp.append(int(input('Enter number')))
        res18 = sum(temp)/len(temp)
        print("mean(x)= ", res18)

    elif op == 19:
        temp = []
        n = int(input("How many numbers you want to use to calculate median?"))
        for i in range(n):
            temp.append(int(input("Enter number")))
        from numpy import median
        res19 = median(temp)
        print("median()= ", res19)

    elif op == 20:
        temp = []
        n = int(input("How many numbers you want to use to calculate mode?"))
        for i in range(n):
            temp.append(int(input("Enter number")))
        from scipy.stats import mode
        res20 = mode(temp)
        print("mode()= ", res20)


    else:
        print('choose another operation')

    new = int(input("do what you want to continue - (yes - 1) , (NO - 0) "))
    if new == 1:
        op = int(input("enter operation"))
    elif new == 0:
        print("Thanks for using the scientific calculator")
        break
```
